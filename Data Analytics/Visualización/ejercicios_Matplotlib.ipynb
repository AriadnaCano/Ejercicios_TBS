8. Crea un gr치fico de barras. Pintamos la altura de un grupo de amigos, cada barra representa a una persona. Ana mide 160 cm, Luis mide 180 cm, Pedro mide 175 cm, Sof칤a mide 190 cm, Carmen mide 170 cm. Las barras ser치n verdes.
Consejo: si no vas a tener que manipular especialmente los ejes, es m치s sencillo dejarlos dentro de la figura.

```python
personas = ["Ana", "Luis", "Pedro", "Sof칤a", "Carmen"]
alturas = [160, 180, 175, 190, 170]
fig = plt.figure()
plt.bar(personas,alturas, color = "green")
```

9. Bas치ndote en el gr치fico anterior, escribe encima de cada barra la altura de cada amigo.
Pista: usa un bucle que lea cada barra de barplot = plt.bar(x,y)
bar tiene los m칠todos get_height(), get_x(), get_width()
plt.text(x,y,valor, va='bottom')

```python
for bar in plt.bar(personas,alturas, color = "green"):
    valor_barra = bar.get_height()
    inicio_barra = bar.get_x()
    anchura_barra = bar.get_width()

    plt.text(inicio_barra + (anchura_barra/3), valor_barra + 2, str(valor_barra))
    plt.yticks([])
```

15. Vamos a introducir pandas. Crea un DataFrame con las columnas "Year" de valores 2015, 2016, 2017, 2018, 2019 y la columna "Sold_items_A" de valores 1000, 3500, 4000, 5500, 7000

```python
data = {"Year":[2015, 2016, 2017, 2018, 2019], "Sold_items_A": [1000, 3500, 4000, 5500, 7000], "Sold_items_B": [2000, 3100, 5000, 4000, 6000]}
df = pd.DataFrame(data)
```

```python
plt.style.use('seaborn-white')
fig = plt.figure()
ax = plt.axes()

x = df["Year"]
ax.plot(x, df["Sold_items_A"], color = "green", linestyle = "--", linewidth = 2.5)
plt.title("Ejercicio 16")
plt.xlabel("A침os")
plt.ylabel("Ventas")

x = df["Year"]
ax.plot(x, df["Sold_items_B"], color = "red", linestyle = ":", linewidth = 2.5)

```
